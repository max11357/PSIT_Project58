```python
from plotly import __version__
from plotly.graph_objs import *
from plotly.offline import download_plotlyjs, init_notebook_mode, iplot
import plotly.plotly as py
init_notebook_mode()

trace1 = Pie(
    domain=dict(
        x=[0, 0.4],
        y=[0.4, 1]
    ),
    hole=0.4,
    hoverinfo='label+percent+name',
    labels=['ENGINEERING', 'BIOLOGY', 'PHYSICS', 'COMPUTER SCIENCE', 'MATHEMATICS', 'LAW', 'CHEMISTRY', 'OTHERS'],
    name='Faculty 55',
    values=[18.20, 11.32, 8.43, 6.03, 5.73, 5.36, 5.07, 39.86]
)

trace2 = Pie(
    domain=dict(
        x=[0.05, 1],
        y=[0.4, 1]
    ),
    hole=0.4,
    hoverinfo='label+percent+name',
    labels=['ENGINEERING', 'BIOLOGY', 'PHYSICS', 'COMPUTER SCIENCE', 'MATHEMATICS', 'LAW', 'CHEMISTRY', 'OTHERS'],
    name='Faculty 56',
    values=[17.25, 11.50, 8.79, 6.36, 6.19, 5.26, 5.18, 39.47]
)

trace3 = Pie(
    domain=dict(
        x=[0.7, 1],
        y=[0.4, 1]
    ),
    hole=0.4,
    hoverinfo='label+percent+name',
    labels=['ENGINEERING', 'BIOLOGY', 'PHYSICS', 'COMPUTER SCIENCE', 'MATHEMATICS', 'LAW', 'CHEMISTRY', 'OTHERS'],
    name='Faculty 57',
    values=[15.97, 10.36, 7.70, 6.70, 5.70, 5.62, 5.28, 42.67]
)


data = Data([trace1, trace2, trace3])

layout = Layout(
    annotations=Annotations([Annotation(x=0.17, y=0.72, font=Font(size=12), showarrow=False, text='Fac 55'),
                                         Annotation(x=0.52, y=0.72, font=Font(size=12), showarrow=False, text='Fac 56'),
                                         Annotation(x=0.89, y=0.72, font=Font(size=12), showarrow=False, text='Fac 57')
                            ]),
    autosize=True,
    height=474,
    hovermode=False,
    title='The most popular Faculty',
    width=1000
)
fig = Figure(data=data, layout=layout)
iplot(fig)
url = py.plot(fig, filename='The most popular Faculty')
```
