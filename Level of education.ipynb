```python
from plotly import __version__
from plotly.graph_objs import *
from plotly.offline import download_plotlyjs, init_notebook_mode, iplot
import csv
init_notebook_mode()

file=open('Level of education.csv')
data=csv.reader(file)
data_list=[row for row in data]
country=[coun[0] for coun in data_list]
total=[coun[1] for coun in data_list]

trace1 = Pie(
    domain=dict(x=[0, 0.51]),
    hole=0.4,
    hoverinfo='label+percent+name',
    labels=country,
    name='Level of education stats',
    values=total
)



data = Data([trace1])

layout = Layout(
    annotations=Annotations([Annotation(x=0.2, y=0.5, font=Font(size=12), showarrow=False, text='Level of EDU 55')]),
    autosize=True,
    height=474,
    hovermode=False,
    title='Level of education 2555 - 2557',
    width=1000
)
fig = Figure(data=data, layout=layout)
iplot(fig)
```
